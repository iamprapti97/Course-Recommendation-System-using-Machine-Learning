COLLABORATIVE FILTERING BASED RECOMMENDER SYSTEM

```python
pt=course.pivot_table(index='course name',columns='subject',values='rating')
```

```python
pt.fillna(0,inplace=True)#810 dimensional space(vector)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_score=cosine_similarity(pt)
similarity_score[0]
```

```python
def recommend(course_name):
    index=np.where(pt.index==course_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    data=[]
    for i in similar_items:
        item=[]
        temp_df=course[course['course name']==pt.index[i[0]]]
        item.extend(list(temp_df['course name'].values))
        item.extend(list(temp_df['instructor'].values))
        item.extend(list(temp_df['image'].values))
        item.extend(list(temp_df['rating'].values))
        item.extend(list(temp_df['subject'].values))
        item.extend(list(temp_df['description'].values))
        data.append(item)
        
    return data
        
```

```python
recommend('100 Days of Code: The Complete Python Pro Bootcamp for 2023')
```
